```python
import pandas as pd
```

```python
df = pd.read_csv("clean.csv")
```

```python
def rating_category(rating):
    if pd.isna(rating):
        return "unknown"
    elif rating >= 7:
        return "positive"
    elif rating <= 4:
        return "negative"
    else:
        return "neutral"
```

```python
def get_random_known_samples(df, n=50, class_col="class"):
    filtered = df[df[class_col] != "unknown"]
    return filtered.sample(n=min(n, len(filtered)), random_state=None)
```

```python
df["label"] = df["rating"].apply(rating_category)
df.shape
```

```python
sample = get_random_known_samples(df, n=50)
print(len(sample))
sample.head()
```
