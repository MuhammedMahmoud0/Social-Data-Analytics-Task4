```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("tmdb_reviews_fast.csv")
```

```python
def rating_category(rating):
    if pd.isna(rating):
        return "unknown"
    elif rating >= 7:
        return "positive"
    elif rating <= 4:
        return "negative"
    else:
        return "neutral"
```

```python
df["label"] = df["rating"].apply(rating_category)
df.shape
```

```python
len(df[df["label"] == "unknown"]["review"])
```

```python
# remove unknown samples
known_df = df[df["label"] != "unknown"]
print(known_df.shape)
```

```python
known_df["label"].value_counts()
```

```python
known_df.head(20)
```

```python
known_df.to_csv("tmdb_reviews_labeled.csv", index=False)
```

```python
from sklearn.utils.class_weight import compute_class_weight

weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(known_df["label"]),
    y=known_df["label"]
)
weights
```

```python
min_class = known_df['label'].value_counts().min()

df_balanced = (
    known_df.groupby('label', group_keys=False)
    .sample(n=min_class, random_state=42)
)

df_balanced["label"].value_counts()
```
